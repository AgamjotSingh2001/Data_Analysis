```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Create a list to store sample means
means=[]
```

```python
#repeat experiment 100 times
for i in range(100):
    #generate a random array of 5 values, with values between 0 and 1
    arr=np.random.rand(5)
    #calculate mean of random sample
    s_mean = np.mean(arr)
    #add S-mean to list
    means.append(s_mean)
```

```python
#plot it along with the mean of the  distribution
sns.histplot(means,kde=True, #kernel desnisty estimation smoothens the plot
            bins=100, color='darkblue')
plt.show()
```

```python
#claculate the mean
mean=sum(means)/len(means)
```

```python
#Plot the mean over the distribution to get a sense of the central tendency 
sns.histplot(means,kde=True, #kernel desnisty estimation smoothens the plot
            bins=100, color='darkblue')
plt.axvline(mean, color='k', linestyle='dashed', linewidth=1)
min_ylim, max_ylim = plt.ylim()
plt.text(mean*1.1, max_ylim*0.9, 'Mean : {:.2f}'.format(mean))
plt.title("Sampling distribution of asmple means of randomly generated samples(no of samples = 100)")
plt.xlabel("Sample mean")
plt.ylabel("Frequency")
plt.show()
```
