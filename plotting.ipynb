# 1. Write a Pandas program to create a line plot of the historical stock prices of Alphabet Inc. between two specific dates.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('alphabet_stock_data.csv')
df.head()
```

```python

df['Date']=pd.to_datetime(df['Date'])
df.info()
df
```

```python
df1=df.set_index('Date',inplace=False)
df
```

```python
plt.plot(df1.iloc[0:100,4])
plt.xlabel("Dates")
plt.ylabel("Stock price at end of the day")
plt.xticks(rotation=45)
plt.show()
```

# 2. Write a Pandas program to create a line plot of the opening, closing stock prices of Alphabet Inc. between two specific dates.

```python
plt.plot(df.iloc[0:100,2],label="Open Stock Price")
plt.plot(df.iloc[0:100,4],label="Closing Stock Price")
plt.xlabel("Dates")
plt.ylabel("Stock price ")
plt.xticks(rotation=45)
plt.legend();
plt.show()
```

# 3. Write a Pandas program to create a bar plot of the trading volume of Alphabet Inc. stock between two specific dates.

```python
df2=df.head()
plt.bar(df2['Date'],df2['Volume'])
plt.xlabel("months")
plt.ylabel("Volume")
plt.xticks(rotation=90)
plt.title("bar")
plt.show()
```

# 4. Write a Pandas program to create a bar plot of opening, closing stock prices of Alphabet Inc. between two specific dates.

```python
plt.bar(df2['Date'],df2['Open'],align="center",width=0.5)
plt.bar(df2['Date'],df2['Close'],align="edge",width=0.5)
plt.xlabel("dates")
plt.ylabel("Stock price")
plt.xticks(rotation=90)
plt.title("bar")
plt.show()
```

# 5. Write a Pandas program to create a stacked bar plot of opening, closing stock prices of Alphabet Inc. between two specific dates.

```python
plt.bar(df2['Date'],df2['Open'],color='r')
plt.bar(df2['Date'],df2['Close'],color='g')
plt.xlabel("dates")
plt.ylabel("Stock price")
plt.xticks(rotation=90)
plt.title("bar")
plt.show()
```

# 6. Write a Pandas program to create a horizontal stacked bar plot of opening,closing stock prices of Alphabet Inc. between two specific dates.

```python
plt.barh(df2['Date'],df2['Open'],color='r')
plt.barh(df2['Date'],df2['Close'],color='g')
plt.ylabel("dates")
plt.xlabel("Stock price")
plt.xticks(rotation=90)
plt.title("bar")
plt.show()
```
